```python
import pandas as pd
import numpy as np

import cv2
import matplotlib.pyplot as plt
```

```python
# Reading images
dog = plt.imread('/content/input/dog.jpg')
cat = cv2.imread('/content/input/cat.jpg')
lena = plt.imread('/content/input/lena-bw.png')
dog.shape, cat.shape, lena.shape
```

```python
pd.Series(dog.flatten()).plot(figsize=(6,4), kind='hist', bins=50, title='Distribution of Pixel')
plt.show()
```

```python
# Display images
# Matplotlib reads in RGB, CV2 reads in BGR
dog_cv = cv2.imread('/content/input/dog.jpg')
fig, axs = plt.subplots(1, 2, figsize=(6,6))
axs[0].imshow(dog)
axs[1].imshow(dog_cv)
axs[0].axis('off')
axs[1].axis('off')
axs[0].set_title('Matplotlib image')
axs[1].set_title('CV2 image')
plt.show()
```

```python
dog_cv_rgb = cv2.cvtColor(dog_cv, cv2.COLOR_BGR2RGB)
fig, ax = plt.subplots(figsize=(4,4))
ax.imshow(dog_cv_rgb)
ax.axis('off')
plt.show()
```

```python
#  Display RGB Channel of our images
fig, axs = plt.subplots(1, 3, figsize=(15,5))
axs[0].imshow(dog[:,:,0], cmap='Reds')
axs[1].imshow(dog[:,:,1], cmap='Greens')
axs[2].imshow(dog[:,:,2], cmap='Blues')
axs[0].axis('off')
axs[1].axis('off')
axs[2].axis('off')
axs[0].set_title('Red channel')
axs[1].set_title('Green channel')
axs[2].set_title('Blue channel')
plt.show()
```

```python
# Image manipulation
dog_gray = cv2.cvtColor(dog, cv2.COLOR_RGB2GRAY)
fig, ax = plt.subplots(figsize=(4,4))
ax.imshow(dog_gray, cmap='Greys')
ax.axis('off')
ax.set_title('Grey image')
plt.show()
```

```python
# Resize images
dog_resized = cv2.resize(dog, None, fx=0.2, fy=0.2)
fig, ax = plt.subplots(figsize=(4,4))
ax.imshow(dog_resized)
ax.axis('off')
ax.set_title('Resize image')
plt.show()
```

```python
dog.shape, dog_resized.shape
```

```python
dog_stretched = cv2.resize(dog, (100, 200))
fig, ax = plt.subplots(figsize=(4,4))
ax.imshow(dog_stretched)
ax.axis('off')
ax.set_title('Stretch image')
plt.show()
```

```python
dog_hd = cv2.resize(dog, (1000,710), interpolation=cv2.INTER_CUBIC)
fig, ax = plt.subplots(figsize=(4,4))
ax.imshow(dog_hd)
ax.axis('off')
ax.set_title('High-definition image')
plt.show()
```

```python
# Sharpening image
kernel_sharpening = np.array([[-1,-1,-1],
                        [-1,9,-1],
                        [-1,-1,-1]])
dog_sharpened = cv2.filter2D(dog, -1, kernel_sharpening)
fig, ax = plt.subplots(figsize=(4,4))
ax.imshow(dog_sharpened)
ax.axis('off')
ax.set_title('Sharpen image')
plt.show()
```

```python
# Blurring image
kernel_3x3 = np.ones((3, 3), np.float32) / 20
dog_blurred = cv2.filter2D(dog, -1, kernel_3x3)
fig, ax = plt.subplots(figsize=(4,4))
ax.imshow(dog_blurred)
ax.axis('off')
ax.set_title('Blur image')
plt.show()
```

```python
# Save images
plt.imsave('dog_sharpened.png', dog_sharpened)
cv2.imwrite('dog_blurred.png', dog_blurred)
```
